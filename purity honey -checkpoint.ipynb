```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
dataframe=pd.read_csv("honey-purity-prediction/Honey Purity prediction/honey purity dataset.csv")
dataframe
```

```python
dataframe.info()
```

```python
dataframe.describe()
```

```python
dataframe.isnull().sum()
```

```python
le=LabelEncoder()
dataframe["Pollen_analysis"]=le.fit_transform(dataframe["Pollen_analysis"])
```

```python
dataframe
```

```python
dataframe["Pollen_analysis"].nunique()
```

```python
X=dataframe[['CS','Density','WC','pH','EC','F','G','Pollen_analysis','Viscosity','Price']]
y=dataframe["Purity"]
print(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
```

```python
model1=RandomForestRegressor()
model2=RandomForestClassifier()
model1
```

```python
model2
```

```python
model1.fit(X_train,y_train)
```

```python
y_predict=model1.predict(X_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2s=r2_score
```

```python
r2s(y_predict,y_test)
```

```python
plt.scatter(y_predict,y_test)
```

```python
residuals = y_test - y_predict

plt.figure(figsize=(8,6))
sns.histplot(residuals, bins=30, kde=True)
plt.axvline(0, color='red', linestyle='--')  # Zero residual line
plt.xlabel("Residual (Error)")
plt.ylabel("Frequency")
plt.title("Residual Distribution")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=100,    # Reduce the number of trees
    max_depth=10,        # Limit tree depth to prevent overfitting
    min_samples_split=5, # Require at least 5 samples to split a node
    random_state=42
)
model
```

```python
model.fit(X_train,y_train)
```

```python
y_p=model.predict(X_test)
y_p
```

```python
y_test
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_predict, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')  # Perfect fit line
plt.xlabel("Actual Purity")
plt.ylabel("Predicted Purity")
plt.title(f"Actual vs Predicted (RÂ² = {r2_score(y_test, y_predict):.2f})")
plt.show()
```
